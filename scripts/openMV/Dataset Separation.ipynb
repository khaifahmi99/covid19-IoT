```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2 as cv
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('labels.csv')
df
```

```python
sns.countplot(df['class'])
```

```python
# target class (associated with label col)

classes = [1, 3]
```

```python
os.mkdir('fall_dataset')
for category in classes:
    os.mkdir('fall_dataset/' + str(category))
```

```python
for category in classes:
    tdf = df[df['class'] == category] # filter dataset BY category
    
    for idx, row in tdf.iterrows():
        name = str(row['index'])
        name = name.zfill(4)
        im = cv.imread('rgb/rgb_' + name + '.png')
        cv2.imwrite('fall_dataset/' + category + '/' + name, im)
```
